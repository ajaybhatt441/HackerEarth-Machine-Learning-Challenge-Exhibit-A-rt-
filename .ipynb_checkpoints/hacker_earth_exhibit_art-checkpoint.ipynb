```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('dataset/train.csv')
test=pd.read_csv('dataset/test.csv')
sample_sub=pd.read_csv('dataset/sample_submission.csv')
```

```python
print(train.shape)
print(test.shape)
print(sample_sub.shape)
```

```python
y=[abs(i) for i in train['Cost']]
```

```python
#'Height', 'Width', 'Weight'
X=train.drop(['Customer Id','Artist Name','Scheduled Date', 'Delivery Date', 'Customer Location','Cost','Material','International', 'Express Shipment', 'Installation Included',
       'Transport', 'Fragile', 'Customer Information', 'Remote Location'],axis=1)
```

```python
categorical=['Material','International', 'Express Shipment', 'Installation Included',
       'Transport', 'Fragile', 'Customer Information', 'Remote Location']
```

```python
X=X.join(pd.get_dummies(train[categorical]))
```

```python
X.shape
```

```python
X.head()
```

```python
X.head()
```

# Data visualization

```python
for i in ['International', 'Express Shipment', 'Installation Included','Transport', 'Fragile', 'Customer Information', 'Remote Location']:
    
    plt.figure()
    sns.countplot(train[i])
       
```

```python
X.isnull().sum() 
```

```python
null_values=[i for i in X.columns if X[i].isnull().sum()>0]
```

```python
null_values
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute=SimpleImputer(strategy='mean').fit(X)
```

```python
Xt=impute.transform(X)
```

```python
Xt
```

```python
 Xt=pd.DataFrame(Xt,columns=X.columns)
```

```python
Xt
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.neighbors import KNeighborsRegressor
```

```python
X_train,X_test,y_train,y_test=train_test_split(Xt,y)
```

```python
gdX=GridSearchCV(XGBRegressor(objective='reg:squarederror'),{'learning_rate':[0.01,0.1,1,.5],'max_depth':[1,2,3,5],'n_estimators':[10,20,50]},cv=5).fit(X_train,y_train)
```

```python
gdX.best_params_
```

```python
gdX.score(X_test,y_test)
```

```python
mean_squared_error(gdX.predict(X_test),y_test)
```

```python
gdn=GridSearchCV(KNeighborsRegressor(),{'n_neighbors':[15,19,21,23]},cv=5).fit(X_train,y_train)
```

```python
gdr=GridSearchCV(RandomForestRegressor(),{'n_estimators':[50,80,100,150],'max_depth':[3,5,7,9,12]},cv=5).fit(X_train,y_train)
```

```python
print('on whole data')
score = 100*max(0, 1-metrics.mean_squared_log_error(gdn.predict(Xt),y))
score1= 100*max(0, 1-metrics.mean_squared_log_error(gdX.predict(Xt),y))
score2 = 100*max(0, 1-metrics.mean_squared_log_error(gdr.predict(Xt),y))
print(score)
print(score1)
print(score2)

print('test')
score = 100*max(0, 1-metrics.mean_squared_log_error(gdn.predict(X_test),y_test))
score1= 100*max(0, 1-metrics.mean_squared_log_error(gdX.predict(X_test),y_test))
score2 = 100*max(0, 1-metrics.mean_squared_log_error(gdr.predict(X_test),y_test))
print(score)
print(score1)
print(score2)
```

```python
print(gdn.best_params_)
print(gdr.best_params_)
```

```python
print(gdn.best_estimator_.score(X_test,y_test))
print(gdr.best_estimator_.score(X_test,y_test))
```

```python
print(mean_squared_error(gdr.predict(X_test),y_test))
print(mean_squared_error(gdn.predict(X_test),y_test))
```
